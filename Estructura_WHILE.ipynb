```python
num = int(input("Ingrese número: "))
cubo= num * num * num

while num > 0:
    cubo = num * num * num
    print("El cubo de", num, "es", cubo)
    num = int(input("Ingrese otro número: "))
print("Finaliza programacion.")
```
