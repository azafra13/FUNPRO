```python
def contador():
  cf=[]
  cantidad=int(input("Ingrese la cantidad de cf: "))
  for i in range(cantidad):
    nota=float(input(f"Ingrese las notas {i+1}: "))
    cf.append(nota)
    print(cf)
  c=0
  for a in cf:
    if a>=3:
      c+=1
      print(f"van {c} cf")
  print(f"{c} fueron las cf mayores o iguales 3")
  return cantidad,cf

def promedio(cf,cantidad):
  totnotas=sum(cf)
  promedio=totnotas/cantidad
  print(f"El promedio es: {promedio}")

a,b=contador ()
promedio(b,a)
```
