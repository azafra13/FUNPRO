#FUNCIONES
Estructura de programación que permite "cortar" un código grande en partes pequeñas y funcionales.
sintaxis de una función


```
def nombre_de_la_función(d,d1,d2,d3,...,dn):
  operación1
  operación2
  función1
  estructura de control
  estructura de selección
  return variable
```
donde:

* d,d1,d2: son los parámetros de la función
* Por dentro puede ir todas las operaciones
* return vuelve una variable local--->global

Existen 4 tipos de funciones
1. Sin parámetro, sin retorno.


```
def función1():
  operación1
  operación2
```
2. Sin parámetro, con retorno.


```
def función2():
  operación1
  operación2
  return dato
```
3. Con parámetro, sin retorno.


```
def función3(color):
  operación1
  operación2
```
4. Con parámetro, con retorno.


```
def función(n1,n2,n3,n4):
  operación1
  operación2
  return promedio
```


````python
#Ejemplo: Desarrolar un programa que de n notas (donde n es mayor que 1), me diga cuál es la mayor nota y cuál es la menor.
notas=int(input("Ingrese la cantidad de notas:"))
nota=[]
for i in range(notas):
  n=int(input(f"Ingrese la nota{i+1}:"))
  nota.append(n)
print(nota)
````

````python
nota_max=max(nota)
nota_min=min(nota)
print(f"La nota máxima es {nota_max} y la peorsita es {nota_min}")
````

````python
def fnotas():
  fnotas=int(input("Ingrese la cantidad de notas:"))
  fnota=[]
  for i in range(fnotas):
   fn=float(input(f"Ingrese la nota {i+1}:"))
   fnota.append(fn)
  return fnota
````

````python
#Función de tipo
def max_min(notas):
  nota_max=max(notas)
  nota_min=min(notas)
  print(f"La nota máxima es {nota_max} y la peorsita es {nota_min}")
````
